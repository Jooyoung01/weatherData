# How to get weather station ID(FIPS) for using NOAA API

```python
### station_id = "TargetID"
### ex: station_id = "CITY:US060013" Los Angeles(LAX) [FIPS code for NOAA_GHCND dataset]
```

```python
folder_path = "your_base_folder_path"
```

### Federal... just for checking api

```python
### US FIPS(Federal..)

import requests
import csv

def search_noaa_stations(file_name, api_key, datasetid="GHCND", locationid="FIPS:US", limit=1000): ##### You can chage the default limit
        url = "https://www.ncdc.noaa.gov/cdo-web/api/v2/stations"
        headers = {"token": api_key}
        params = {
            "datasetid": datasetid,  ##### "GHCND" (Global Historical Climatology Network - Daily)
            "locationid": locationid,  ##### "FIPS:US" (USA) (FIPS Code_ ex: "CITY:US060013" Los Angeles(LAX))
            "limit": limit
        }
        with open(file_name, mode='w', newline='', encoding='utf-8') as file:
            writer = csv.writer(file)
            writer.writerow(["Station_ID", "Name", "Latitude", "Longitude"])  # Writing header
            response = requests.get(url, headers=headers, params=params)
            stations = response.json()
            for station in stations['results']:
                print(f"Station ID: {station['id']}, Name: {station['name']}, Latitude: {station['latitude']}, Longitude: {station['longitude']}")
                writer.writerow([station['id'], station['name'], station['latitude'], station['longitude']])

# NOAA API Key
# api_key = "YOUR_API_KEY_HERE"

file_name = f"{folder_path}/noaa_stations.csv"
search_noaa_stations(file_name,api_key)
```

## GET FIPS Code from NOAA

### COUNTY LEVEL

```python
import requests
import csv

# api_key = "YOUR_API_KEY_HERE"
url = "https://www.ncdc.noaa.gov/cdo-web/api/v2/locations"
headers = {"token": api_key}
params = {"locationcategoryid": "CNTY", "limit": 1000} ## FIND LA COUNTY!

response = requests.get(url, headers=headers, params=params)
locations = response.json()

with open(f'{folder_path}/get_FIPS_County.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["Location_NAME", "FIPS_code"])
    for location in locations.get('results', []):
        print(f"Location ID: {location['id']}, Name: {location['name']}")
        writer.writerow([location['name'], location['id']])

```

## Let's get LA CITY Weather stations

```python
import requests

def search_la_noaa_stations(api_key):
    url = "https://www.ncdc.noaa.gov/cdo-web/api/v2/stations"
    headers = {"token": api_key}
    params = {
        "datasetid": "GHCND",  # Global Historical Climatology Network - Daily dataset
        "locationid": "CITY:US060013",  # LA FIPS Code
        "limit": 100
    }

    response = requests.get(url, headers=headers, params=params)
    stations = response.json()

    for station in stations.get('results', []):
        print(f"Station ID: {station['id']}, Name: {station['name']}")

# api_key = "YOUR_API_KEY_HERE"
search_la_noaa_stations(api_key)
```
